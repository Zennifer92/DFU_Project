```r
 for (file_name in file_list1) {
  #load data
  G1A<- read.csv(file_name)
  str(G1A)
  class(G1A)
  G1A
  G1A_matrix <- data.matrix(G1A)
     
     # Initialize the Seurat object with the raw (non-normalized data).
  G1A.seurat.obj <- CreateSeuratObject(counts = G1A_matrix, project = "G1A_matrix", min.cells = 3, min.features = 200)
  str(G1A.seurat.obj)
  G1A.seurat.obj
  
  # 1. QC -------
  View(G1A.seurat.obj@meta.data)
  # % MT reads
  G1A.seurat.obj[["percent.mt"]] <- PercentageFeatureSet(G1A.seurat.obj, pattern = "^MT-")
  View(G1A.seurat.obj@meta.data)
  
  VlnPlot(G1A.seurat.obj, features = c("nFeature_RNA", "nCount_RNA", "percent.mt"), ncol = 3)
  FeatureScatter(G1A.seurat.obj, feature1 = "nCount_RNA", feature2 = "nFeature_RNA") +
    geom_smooth(method = 'lm')
  # 2. Filtering -----------------
  G1A.seurat.obj <- subset(G1A.seurat.obj, subset = nFeature_RNA > 200 & nFeature_RNA < 2500 & 
                             percent.mt < 5)
  
  # 3. Normalize data ----------
  #nsclc.seurat.obj <- NormalizeData(nsclc.seurat.obj, normalization.method = "LogNormalize", scale.factor = 10000)
  # OR
  G1A.seurat.obj <- NormalizeData(G1A.seurat.obj)
  str(G1A.seurat.obj)
  
  
  # 4. Identify highly variable features --------------
  G1A.seurat.obj <- FindVariableFeatures(G1A.seurat.obj, selection.method = "vst", nfeatures = 2000)
  
  # Identify the 10 most highly variable genes
  top10 <- head(VariableFeatures(G1A.seurat.obj), 10)
  
  # plot variable features with and without labels
  plot1 <- VariableFeaturePlot(G1A.seurat.obj)
  LabelPoints(plot = plot1, points = top10, repel = TRUE)
  
  
  # 5. Scaling -------------
  all.genes <- rownames(G1A.seurat.obj)
  G1A.seurat.obj <- ScaleData(G1A.seurat.obj, features = all.genes)
  
  str(G1A.seurat.obj)
  
  # 6. Perform Linear dimensionality reduction --------------
  G1A.seurat.obj <- RunPCA(G1A.seurat.obj, features = VariableFeatures(object = G1A.seurat.obj))
  
  # visualize PCA results
  print(G1A.seurat.obj[["pca"]], dims = 1:5, nfeatures = 5)
  DimHeatmap(G1A.seurat.obj, dims = 1, cells = 500, balanced = TRUE)
  
  
  # determine dimensionality of the data
  ElbowPlot(G1A.seurat.obj)
  
  
  # 7. Clustering ------------
  G1A.seurat.obj <- FindNeighbors(G1A.seurat.obj, dims = 1:15)
  
  # understanding resolution
  G1A.seurat.obj <- FindClusters(G1A.seurat.obj, resolution = c(0.1,0.3, 0.5, 0.7, 1))
  View(G1A.seurat.obj@meta.data)
  
  DimPlot(G1A.seurat.obj, group.by = "RNA_snn_res.0.5", label = TRUE)
  
  # setting identity of clusters
  Idents(G1A.seurat.obj)
  Idents(G1A.seurat.obj) <- "RNA_snn_res.0.1"
  Idents(G1A.seurat.obj)
  
  # non-linear dimensionality reduction --------------
  # If you haven't installed UMAP, you can do so via reticulate::py_install(packages =
  # 'umap-learn')
  G1A.seurat.obj <- RunUMAP(G1A.seurat.obj, dims = 1:15)
  # note that you can set `label = TRUE` or use the LabelClusters function to help label
  # individual clusters
  DimPlot(G1A.seurat.obj, reduction = "umap")
  
  data_list1[file_name] <-G1A.seurat.obj
  
}
data_list1
```
